## Tasks

Given historical data on loans given out with information on whether or not the borrower defaulted (charge-off), can we build a model thatcan predict wether or nor a borrower will pay back their loan? This way in the future when we get a new potential customer we can assess whether or not they are likely to pay back the loan.

```python
# total_acc feature correlates with the mort_acc

# calculate the mean value for the mort_acc per total_acc entry
print("Mean of mort_acc column per total_acc")
df.groupby('total_acc').mean()['mort_acc']
```

```python
# fill in the missing mort_acc values based on their total_acc value
# If the mort_acc is missing, then we will fill in that missing value with the mean value 
# corresponding to its total_acc value

# pd Series group by total_acc with mean mort_acc value
total_acc_avg = df.groupby('total_acc').mean()['mort_acc']
total_acc_avg
```

```python
def fill_mort_acc(total_acc,mort_acc):
    '''
    Accepts the total_acc and mort_acc values for the row.
    Checks if the mort_acc is NaN , if so, it returns the avg mort_acc value
    for the corresponding total_acc value for that row.
    
    total_acc_avg here should be a Series or dictionary containing the mapping of the
    groupby averages of mort_acc per total_acc values.
    '''
    if np.isnan(mort_acc):
        return total_acc_avg[total_acc]
    else:
        return mort_acc
```

```python
df['mort_acc'] = df.apply(lambda x: fill_mort_acc(x['total_acc'], x['mort_acc']), axis=1)
```

```python
# checking
df.isnull().sum()
```

```python
#4 REVOL_UTIL
#6 PUB REC BANKRUPTCIES

# revol_util and the pub_rec_bankruptcies have missing data points, 
# but they account for less than 0.5% of the total data, so we just dropna
df = df.dropna()

# checking
df.isnull().sum()
```

# Prediction Keras API

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X = df2.drop('loan_repaid', axis = 1).values
y = df2['loan_repaid'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
scaler = MinMaxScaler()
```

```python
# Scaling
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
# Create a model
model = Sequential()

model.add(Dense(85, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(40, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(20, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(10, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(5, activation = 'relu'))
model.add(Dropout(0.2))

# BINARY CLASSIFICATION
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss='binary_crossentropy', optimizer = 'adam')
```

```python
# Stop training when a monitored metric has stopped improving.
early_stop = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 8)
```

```python
# Fitting
model.fit(x = X_train, y = y_train, epochs = 40, batch_size=128, validation_data=(X_test, y_test),
         callbacks = [early_stop])
```

```python
# Compere loss and val_loss - avoid overfitting
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```

<p>The accuracy rate is 89%. However, this result is not very good, because here we have a case where the data is not balanced. Therefore, we should focus more on the f1 score.</p>
<br>
<p>Class 1 fared very well at 93%, mainly because 80% of the data includes this class - 1. In the case of class 0, the f1 score is 62%, which is not very well, but not too bad with these unbalanced data.</p>
<br>
<p>Let's try to use SMOTE to balance the data and reuse the model to see if better results can be achieved, remembering not to overfitting.</p>

## Dealing with unbalanced data

```python
################################ SMOTE ########################################
```

```python
from imblearn.over_sampling import SMOTE

# Using smote to minority class
smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_sample(X, y)
```

```python
# new train_test_split after using smote
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.33)
```

```python
# Scaling
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Create the same model
model = Sequential()

model.add(Dense(85, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(40, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(20, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(10, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(5, activation = 'relu'))
model.add(Dropout(0.2))

# BINARY CLASSIFICATION
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss='binary_crossentropy', optimizer = 'adam')
```

```python
# Fitting
model.fit(x = X_train, y = y_train, epochs = 40, batch_size=128, validation_data=(X_test, y_test),
         callbacks = [early_stop])
```

```python
# Compare loss and val_loss
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
predictions = model.predict_classes(X_test)
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test, predictions))
```

Thanks to the use of smote, we were able to improve our results.
